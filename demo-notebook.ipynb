```python
# rename column name
df.rename(columns = {'Title':'Movie_Title'}, inplace = True)
df.head()
```

```python
# derive new column in pandas dataframe
df["Runtime (Hours)"] = round(df[["Runtime (Minutes)"]]/60,2)
df.head()
```

```python
# delete column from dataframe
del df["Rank"]
```

```python
df.head()
```

```python
# filter condition in pandas
years = [2014,2016]
rslt_df = df[(df['Runtime (Hours)'] > 2) & df['Year'].isin(years)]
rslt_df.head()
```

```python
# arrange dataframe in ascending or descenting
df.sort_values(by=['Year','Rating'], ascending=[True,False],inplace=True)
df.head()
```

```python
# group by in pandas
df2 =df.groupby(['Year']).sum()
df2.head()
```

```python
# group by in pandas
df2 = df.groupby(['Year','Genre']).sum()
df2.to_csv('/kaggle/working/result.csv', sep=',', encoding='utf-8')
```
